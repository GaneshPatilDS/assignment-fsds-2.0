**Question 1:**
```python
class DivisibleBySeven:
    def __init__(self, n):
        self.n = n

    def divisible_by_seven_generator(self):
        for num in range(self.n + 1):
            if num % 7 == 0:
                yield num

n = int(input("Enter the value of n: "))
divisible_obj = DivisibleBySeven(n)
for num in divisible_obj.divisible_by_seven_generator():
    print(num, end=" ")
```

**Question 2:**
```python
def word_frequency(input_string):
    words = input_string.split()
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    sorted_frequency = dict(sorted(frequency.items()))
    return sorted_frequency

input_string = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."
output_result = word_frequency(input_string)
for key, value in output_result.items():
    print(f"{key}:{value}")
```

**Question 3:**
```python
class Person:
    def getGender(self):
        return "Unknown"

class Male(Person):
    def getGender(self):
        return "Male"

class Female(Person):
    def getGender(self):
        return "Female"

male_obj = Male()
female_obj = Female()
print(male_obj.getGender())
print(female_obj.getGender())
```

**Question 6:**
```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

sorted_list = [2, 5, 7, 10, 15, 20, 25, 30]
target_element = 15
result_index = binary_search(sorted_list, target_element)
print(f"Element {target_element} found at index {result_index}")
```
