1. Set the variable `test1` to the string 'This is a test of the emergency text system,' and save `test1` to a file named `test.txt`.

```python
test1 = 'This is a test of the emergency text system,'
with open('test.txt', 'w') as file:
    file.write(test1)
```

2. Read the contents of the file `test.txt` into the variable `test2`. Is there a difference between `test1` and `test2`?

```python
with open('test.txt', 'r') as file:
    test2 = file.read()

print(test1 == test2)  # True if the contents are the same
```

4. Use the `sqlite3` module to create a SQLite database called `books.db`, and a table called `books` with these fields: `title (text)`, `author (text)`, and `year (integer)`.

```python
import sqlite3

conn = sqlite3.connect('books.db')
cursor = conn.cursor()

cursor.execute('''CREATE TABLE IF NOT EXISTS books
                (title TEXT, author TEXT, year INTEGER)''')

conn.commit()
conn.close()
```
------------------------------------------------------------------------------------------------------------------------------------------------------

5. Read `books.csv` and insert its data into the `book` table.

```python
import csv

with open('books.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header row
    data_to_insert = [tuple(row) for row in csv_reader]

conn = sqlite3.connect('books.db')
cursor = conn.cursor()

cursor.executemany('INSERT INTO books VALUES (?, ?, ?)', data_to_insert)

conn.commit()
conn.close()
```
------------------------------------------------------------------------------------------------------------------------------------------------------


6. Select and print the `title` column from the `book` table in alphabetical order.

```python
conn = sqlite3.connect('books.db')
cursor = conn.cursor()

cursor.execute('SELECT title FROM books ORDER BY title')
titles = cursor.fetchall()

for title in titles:
    print(title[0])

conn.close()
```
------------------------------------------------------------------------------------------------------------------------------------------------------

7. From the `book` table, select and print all columns in the order of publication.

```python
conn = sqlite3.connect('books.db')
cursor = conn.cursor()

cursor.execute('SELECT * FROM books ORDER BY year')
books = cursor.fetchall()

for book in books:
    print(book)

conn.close()
```

8. Use the `sqlalchemy` module to connect to the SQLite database `books.db` that you just made in exercise 4.

```python
from sqlalchemy import create_engine, Table, Column, String, Integer, MetaData

engine = create_engine('sqlite:///books.db')
metadata = MetaData()

books = Table('books', metadata,
              Column('title', String),
              Column('author', String),
              Column('year', Integer)
              )

metadata.create_all(engine)
```

9. Install the Redis server and the Python redis library (`pip install redis`) on your computer. Create a Redis hash called `test` with the fields `count (1)` and `name ('Fester Bestertester')`. Print all the fields for `test`.

```python
import redis

redis_client = redis.StrictRedis(host='localhost', port=6379, db=0)
redis_client.hset('test', 'count', 1)
redis_client.hset('test', 'name', 'Fester Bestertester')

fields = redis_client.hgetall('test')
print(fields)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

10. Increment the `count` field of `test` and print it.

```python
redis_client.hincrby('test', 'count', 1)
count = redis_client.hget('test', 'count')
print(count)
```
