````python
Sure, here are Python programs for each of the tasks you mentioned:

1. **Add Two Matrices:**
```python
def add_matrices(matrix1, matrix2):
    result = [[matrix1[i][j] + matrix2[i][j] for j in range(len(matrix1[0]))] for i in range(len(matrix1))]
    return result

def input_matrix(rows, columns):
    matrix = []
    for i in range(rows):
        row = [int(x) for x in input(f"Enter space-separated elements of row {i+1}: ").split()]
        matrix.append(row)
    return matrix

rows = int(input("Enter the number of rows: "))
columns = int(input("Enter the number of columns: "))

print("Enter elements of the first matrix:")
matrix1 = input_matrix(rows, columns)

print("Enter elements of the second matrix:")
matrix2 = input_matrix(rows, columns)

result = add_matrices(matrix1, matrix2)
print("Sum of the two matrices:")
for row in result:
    print(row)
```

2. **Multiply Two Matrices:**
```python
def multiply_matrices(matrix1, matrix2):
    rows1, columns1 = len(matrix1), len(matrix1[0])
    rows2, columns2 = len(matrix2), len(matrix2[0])

    if columns1 != rows2:
        print("Matrices cannot be multiplied.")
        return None

    result = [[0 for _ in range(columns2)] for _ in range(rows1)]

    for i in range(rows1):
        for j in range(columns2):
            for k in range(columns1):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result

def input_matrix(rows, columns):
    matrix = []
    for i in range(rows):
        row = [int(x) for x in input(f"Enter space-separated elements of row {i+1}: ").split()]
        matrix.append(row)
    return matrix

rows1 = int(input("Enter the number of rows for the first matrix: "))
columns1 = int(input("Enter the number of columns for the first matrix: "))

rows2 = int(input("Enter the number of rows for the second matrix: "))
columns2 = int(input("Enter the number of columns for the second matrix: "))

print("Enter elements of the first matrix:")
matrix1 = input_matrix(rows1, columns1)

print("Enter elements of the second matrix:")
matrix2 = input_matrix(rows2, columns2)

result = multiply_matrices(matrix1, matrix2)
if result:
    print("Resultant matrix after multiplication:")
    for row in result:
        print(row)
```

3. **Transpose a Matrix:**
```python
def transpose_matrix(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

def input_matrix(rows, columns):
    matrix = []
    for i in range(rows):
        row = [int(x) for x in input(f"Enter space-separated elements of row {i+1}: ").split()]
        matrix.append(row)
    return matrix

rows = int(input("Enter the number of rows: "))
columns = int(input("Enter the number of columns: "))

print("Enter elements of the matrix:")
matrix = input_matrix(rows, columns)

transposed_matrix = transpose_matrix(matrix)
print("Transposed matrix:")
for row in transposed_matrix:
    print(row)
```

4. **Sort Words in Alphabetic Order:**
```python
def sort_words_alphabetic(sentence):
    words = sentence.split()
    words.sort()
    return " ".join(words)

sentence = input("Enter a sentence: ")
result = sort_words_alphabetic(sentence)
print("Sorted words in alphabetic order:")
print(result)
```

5. **Remove Punctuation From a String:**
```python
import string

def remove_punctuation(input_string):
    return input_string.translate(str.maketrans('', '', string.punctuation))

text = input("Enter a string: ")
result = remove_punctuation(text)
print("String without punctuation:")
print(result)
```

The last program removes all punctuation marks from the input string using the `str.translate()` method along with `str.maketrans()`.
````
