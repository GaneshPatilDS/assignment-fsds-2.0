1. Add the current date to the text file `today.txt` as a string.

```python
from datetime import date

current_date = str(date.today())

with open('today.txt', 'w') as file:
    file.write(current_date)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

2. Read the text file `today.txt` into the string `today_string`.

```python
with open('today.txt', 'r') as file:
    today_string = file.read()

print(today_string)
```
-----------------------------------------------------------------------------------------------------------------------------------------------------

3. Parse the date from `today_string`.

```python
parsed_date = date.fromisoformat(today_string)
print(parsed_date)
```

6. Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between one and five, print the current time, and then exit.

```python
import multiprocessing
import time
import random
from datetime import datetime

def print_current_time():
    wait_time = random.randint(1, 5)
    time.sleep(wait_time)
    current_time = datetime.now().strftime('%H:%M:%S')
    print(f"Process {multiprocessing.current_process().name} - Current Time: {current_time}")

processes = []
for i in range(3):
    process = multiprocessing.Process(target=print_current_time)
    processes.append(process)
    process.start()

for process in processes:
    process.join()
```

9. When will you be (or when were you) 10,000 days old?

```python
from datetime import timedelta

days_to_add = 10000
thousand_days_old = day_of_birth + timedelta(days=days_to_add)
print(thousand_days_old)
```
