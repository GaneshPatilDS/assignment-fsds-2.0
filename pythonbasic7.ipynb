# ANS :

1) The feature responsible for generating Regex objects is called the "re" module.
----------------------------------------------------------------------------------------------------------------------------------------

2) Raw strings are often used in Regex objects to avoid the need for excessive escaping. They treat backslashes as literal characters instead of escape characters, making it easier to write and read regular expressions.
-----------------------------------------------------------------------------------------------------------------------------------------

3) The search() method returns a Match object if a match is found. Otherwise, it returns None.
-----------------------------------------------------------------------------------------------------------------------------------------

4) To retrieve the actual strings that match the pattern from a Match object, you can use the group() method.
-----------------------------------------------------------------------------------------------------------------------------------------

5) In the given regex, group zero (group 0) covers the entire match, while group 1 covers the first set of parentheses (\d\d\d) and group 2 covers the second set of parentheses (\d\d\d-\d\d\d\d).
------------------------------------------------------------------------------------------------------------------------------------------
6) To indicate that you want a regex to fit real parentheses and periods, you need to escape them using a backslash (). For example, ( and . would represent literal parentheses and periods, respectively.
------------------------------------------------------------------------------------------------------------------------------------------

7) The findall() method returns a list of strings when the regex pattern has no groups defined. However, if the pattern contains groups, it returns a list of string tuples where each tuple represents a match and contains the strings that correspond to each group.
-------------------------------------------------------------------------------------------------------------------------------------------

8) In standard expressions, the | character represents the logical OR operation. It allows you to match either the expression on its left or the expression on its right.
-------------------------------------------------------------------------------------------------------------------------------------------

9) In regular expressions, the dot (.) character (period) is a special metacharacter that matches any character except a newline character.
------------------------------------------------------------------------------------------------------------------------------------------
10)  regular expressions, the "+" character is a quantifier that matches one or more occurrences of the preceding element. The "" character, on the other hand, is a quantifier that matches zero or more occurrences of the preceding element. In summary, "+" requires at least one occurrence, while "" allows for zero or more occurrences.
--------------------------------------------------------------------------------------------------------------------------------------------
11) In regular expressions, {4} specifies exactly 4 occurrences of the preceding element, while {4,5} specifies a range between 4 and 5 occurrences of the preceding element.
--------------------------------------------------------------------------------------------------------------------------------------------

12) In regular expressions, \d signifies a shorthand character class that matches any digit (0-9), \w matches any alphanumeric character (letters, digits, and underscores), and \s matches any whitespace character (spaces, tabs, newlines).
--------------------------------------------------------------------------------------------------------------------------------------------

13) On the other hand, \D represents the negation of \d and matches any non-digit character, \W represents the negation of \w and matches any non-alphanumeric character, and \S represents the negation of \s and matches any non-whitespace character.
--------------------------------------------------------------------------------------------------------------------------------------------

14) The .? is a non-greedy match that matches as few characters as possible, while . is a greedy match that matches as many characters as possible.
---------------------------------------------------------------------------------------------------------------------------------------------


15) To match both numbers and lowercase letters with a character class, you can use the pattern [0-9a-z]. This character class will match any digit from 0 to 9 or any lowercase letter from a to z.
---------------------------------------------------------------------------------------------------------------------------------------------


16) To make a regular expression case-insensitive, you can pass the re.IGNORECASE flag as the second argument to re.compile(). For example, re.compile(pattern, re.IGNORECASE).
---------------------------------------------------------------------------------------------------------------------------------------------

17) The . character normally matches any character except a newline character. However, if re.DOTALL is passed as the second argument in re.compile(), the . character will match any character including newline characters.
----------------------------------------------------------------------------------------------------------------------------------------------

18) If numReg = re.compile(r'\d+'), the numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4 hen') will return the string 'X drummers, X pipers, five rings, X hen'.
-----------------------------------------------------------------------------------------------------------------------------------------------

19) Passing re.VERBOSE as the second argument to re.compile() allows you to add whitespace and comments within the regular expression pattern. It helps in improving the readability and maintainability of complex regular expressions.
-----------------------------------------------------------------------------------------------------------------------------------------------

20) To match a number with commas for every three digits, you can use the regex pattern r'\d{1,3}(,\d{3})*'. This pattern matches one to three digits followed by zero or more occurrences of a comma and three digits.
-----------------------------------------------------------------------------------------------------------------------------------------------


21) To match the full name of someone whose last name is Watanabe with the assumption that the first name is always capitalized and one word, you can use the regex pattern r'[A-Z][a-z]* Watanabe'.
-------------------------------------------------------------------------------------------------------------------------------------------------
22) To match a sentence where the first word is either Alice, Bob, or Carol; the second word is either eats, pets, or throws; the third word is apples, cats, or baseballs; and the sentence ends with a period, you can use the case-insensitive regex pattern r'(Alice|Bob|Carol) (eats|pets|throws) (apples|cats|baseballs).'.
