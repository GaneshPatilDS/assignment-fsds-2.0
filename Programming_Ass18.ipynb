# ANS :



Question 1:
```python
def filter_list(lst):
    return [item for item in lst if isinstance(item, int)]

print(filter_list([1, 2, "a", "b"]))
print(filter_list([1, "a", "b", 0, 15]))
print(filter_list([1, 2, "aasf", "1", "123", 123]))
```

Question 2:
```python
def reverse(string):
    return string[::-1].swapcase()

print(reverse("Hello World"))
print(reverse("ReVeRsE"))
print(reverse("Radar"))
```

Question 4:
```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
print(factorial(3))
print(factorial(1))
print(factorial(0))
```

Question 5:
```python
def move_to_end(lst, target):
    return [item for item in lst if item != target] + [item for item in lst if item == target]

print(move_to_end([1, 3, 2, 4, 4, 1], 1))
print(move_to_end([7, 8, 9, 1, 2, 3, 4], 9))
print(move_to_end(["a", "a", "a", "b"], "a"))
```
