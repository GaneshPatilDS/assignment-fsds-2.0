# Ans:



1. 
```python
def func(a, b=6, c=8):
    print(a, b, c)

func(1, 2)
```
Result: `1 2 8`
Explanation: The function `func()` has three parameters `a`, `b`, and `c`, with default values `6` for `b` and `8` for `c`. When calling `func(1, 2)`, `a` takes the value `1`, `b` takes the value `2`, and `c` retains its default value of `8`.

------------------------------------------------------------------------------------------------------------------------------------------------------
2. 
```python
def func(a, b, c=5):
    print(a, b, c)

func(1, c=3, b=2)
```
Result: `1 2 3`
Explanation: The function `func()` has three parameters `a`, `b`, and `c`, with a default value of `5` for `c`. When calling `func(1, c=3, b=2)`, `a` takes the value `1`, `b` takes the value `2`, and `c` is explicitly set to `3`, overriding its default value.

------------------------------------------------------------------------------------------------------------------------------------------------------

3. 
```python
def func(a, *pargs):
    print(a, pargs)

func(1, 2, 3)
```
Result: `1 (2, 3)`
Explanation: The function `func()` has a parameter `a`, followed by `*pargs`, which allows for a variable number of positional arguments. When calling `func(1, 2, 3)`, `a` takes the value `1`, and `pargs` becomes a tuple containing the remaining positional arguments `(2, 3)`.

------------------------------------------------------------------------------------------------------------------------------------------------------

4. 
```python
def func(a, **kargs):
    print(a, kargs)

func(a=1, c=3, b=2)
```
Result: `1 {'c': 3, 'b': 2}`
Explanation: The function `func()` has a parameter `a`, followed by `**kargs`, which allows for a variable number of keyword arguments. When calling `func(a=1, c=3, b=2)`, `a` takes the value `1`, and `kargs` becomes a dictionary containing the keyword arguments `{'c': 3, 'b': 2}`.

------------------------------------------------------------------------------------------------------------------------------------------------------

5. 
```python
def func(a, b, c=8, d=5):
    print(a, b, c, d)

func(1, *(5, 6))
```
Result: `1 5 6 5`
Explanation: The function `func()` has four parameters `a`, `b`, `c`, and `d`, with default values `8` for `c` and `5` for `d`. When calling `func(1, *(5, 6))`, `a` takes the value `1`, `b` takes the first value from the tuple `(5, 6)` which is `5`, and `c` takes the second value from the tuple, which is `6`. Since no value is provided for `d`, it retains its default value of `5`.

6. 
```python
def func(a, b, c):
    a = 2
    b[0] = 'x'
    c['a'] = 'y'

l = 1
m = [1]
n = {'a': 0}
func(l, m, n)

print(l, m, n)
```
Result: `1 ['x'] {'a': 'y'}`
Explanation: The function `func()` takes three arguments `a`, `b`, and `c`. Inside the function, `a` is assigned the value `2`, which does not affect the outer variable `l` since it is immutable (an integer). However, `b` is a list, and the first element of the list `m` is changed to `'x'`. Similarly, `c` is a dictionary, and the value associated with the key `'a'` in the dictionary `n` is changed to `'y'`. The changes made to the mutable objects `m` and `n` are reflected in the final print statement.
