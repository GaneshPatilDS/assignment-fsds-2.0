1. **Find Sum of Elements in List:**
```python
def sum_of_elements(lst):
    return sum(lst)

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
result = sum_of_elements(input_list)
print("Sum of elements in the list:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

2. **Multiply All Numbers in the List:**
```python
def multiply_elements(lst):
    result = 1
    for num in lst:
        result *= num
    return result

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
result = multiply_elements(input_list)
print("Result of multiplying all numbers in the list:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

3. **Find the Smallest Number in a List:**
```python
def find_smallest_number(lst):
    return min(lst)

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
result = find_smallest_number(input_list)
print("Smallest number in the list:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

4. **Find the Largest Number in a List:**
```python
def find_largest_number(lst):
    return max(lst)

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
result = find_largest_number(input_list)
print("Largest number in the list:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

5. **Find the Second Largest Number in a List:**
```python
def find_second_largest_number(lst):
    if len(lst) < 2:
        return "List should have at least two elements."

    largest = second_largest = float('-inf')
    for num in lst:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    if second_largest == float('-inf'):
        return "There is no second largest number in the list."
    else:
        return second_largest

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
result = find_second_largest_number(input_list)
print("Second largest number in the list:", result)
```

6. **Find N Largest Elements from a List:**
```python
def find_n_largest_elements(lst, n):
    return sorted(lst, reverse=True)[:n]

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
n = int(input("Enter the value of N: "))
result = find_n_largest_elements(input_list, n)
print(f"{n} largest elements in the list:", result)
```

7. **Print Even Numbers in a List:**
```python
def print_even_numbers(lst):
    even_numbers = [num for num in lst if num % 2 == 0]
    return even_numbers

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
result = print_even_numbers(input_list)
print("Even numbers in the list:", result)
```

8. **Print Odd Numbers in a List:**
```python
def print_odd_numbers(lst):
    odd_numbers = [num for num in lst if num % 2 != 0]
    return odd_numbers

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
result = print_odd_numbers(input_list)
print("Odd numbers in the list:", result)
```

9. **Remove Empty Lists from a List:**
```python
def remove_empty_lists(lst):
    return [sublist for sublist in lst if sublist]

input_list = [[1, 2, 3], [], [4, 5], [], [6, 7, 8, 9], []]
result = remove_empty_lists(input_list)
print("List after removing empty lists:", result)
```

10. **Cloning or Copying a List:**
```python
def clone_list(lst):
    return lst.copy()

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
cloned_list = clone_list(input_list)
print("Cloned list:", cloned_list)
```

11. **Count Occurrences of an Element in a List:**
```python
def count_occurrences(lst, element):
    return lst.count(element)

input_list = [int(x) for x in input("Enter space-separated elements of the list: ").split()]
element_to_count = int(input("Enter the element to count: "))
result = count_occurrences(input_list, element_to_count)
print(f"Occurrences of {element_to_count} in the list:", result)
```
