# ANS:



**Question 1:**
```python
import math

def calculate_q(sequence):
    C, H = 50, 30
    result = []
    for D in sequence:
        Q = int(math.sqrt((2 * C * D) / H))
        result.append(Q)
    return result

input_sequence = [int(x) for x in input("Enter comma-separated values of D: ").split(",")]
output_result = calculate_q(input_sequence)
print(",".join(str(x) for x in output_result))
```

**Question 2:**
```python
def generate_2d_array(X, Y):
    result = [[i * j for j in range(Y)] for i in range(X)]
    return result

X, Y = [int(x) for x in input("Enter X and Y separated by commas: ").split(",")]
output_result = generate_2d_array(X, Y)
print(output_result)
```


**Question 3:**
```python
def sort_words_alphabetically(words):
    words_list = words.split(",")
    sorted_words = sorted(words_list)
    return ",".join(sorted_words)

input_words = input("Enter comma-separated words: ")
output_result = sort_words_alphabetically(input_words)
print(output_result)
```

**Question 4:**
```python
def remove_duplicates_and_sort(input_string):
    words = input_string.split()
    unique_words = sorted(set(words))
    return " ".join(unique_words)

input_string = input("Enter whitespace-separated words: ")
output_result = remove_duplicates_and_sort(input_string)
print(output_result)
```

**Question 5:**
```python
def count_letters_and_digits(sentence):
    letters_count = sum(1 for char in sentence if char.isalpha())
    digits_count = sum(1 for char in sentence if char.isdigit())
    return letters_count, digits_count

input_sentence = input("Enter a sentence: ")
letters, digits = count_letters_and_digits(input_sentence)
print("LETTERS", letters)
print("DIGITS", digits)
```

**Question 6:**
```python
import re

def check_valid_passwords(passwords):
    valid_passwords = []
    for password in passwords:
        if (6 <= len(password) <= 12 and
            re.search("[a-z]", password) and
            re.search("[0-9]", password) and
            re.search("[A-Z]", password) and
            re.search("[$#@]", password)):
            valid_passwords.append(password)
    return valid_passwords

input_passwords = input("Enter comma-separated passwords: ").split(",")
output_result = check_valid_passwords(input_passwords)
print(",".join(output_result))
```
