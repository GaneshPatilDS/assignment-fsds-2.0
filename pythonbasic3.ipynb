Q.6)What happens to variables in a local scope when the function call returns?

ANS:



    When a function call returns, the local scope is destroyed, and all the variables defined within that scope are no longer accessible. This is known as "variable scope" in programming, and it is a fundamental concept in many programming languages, including Python. The memory used by these variables is also freed up by the garbage collector.

Q.14)What is the purpose of the try clause? What is the purpose of the except clause?

ANS:

   The purpose of the try clause is to define a block of code that may cause an exception to be raised, and to specify what should happen if an exception is raised.

The except clause is used to handle exceptions that are raised within the try block. The purpose of the except clause is to specify what should happen if a particular exception is raised within the try Block.
