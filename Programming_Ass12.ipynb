# ANS:



1. **Extract Unique Values from Dictionary Values:**
```python
def extract_unique_values(dictionary):
    unique_values = set()
    for value_list in dictionary.values():
        unique_values.update(value_list)
    return unique_values

my_dict = {
    'a': [1, 2, 3],
    'b': [2, 3, 4],
    'c': [4, 5, 6]
}

result = extract_unique_values(my_dict)
print("Unique values from dictionary values:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

2. **Find the Sum of All Items in a Dictionary:**
```python
def sum_of_items(dictionary):
    total_sum = sum(sum(value_list) for value_list in dictionary.values())
    return total_sum

my_dict = {
    'a': [1, 2, 3],
    'b': [2, 3, 4],
    'c': [4, 5, 6]
}

result = sum_of_items(my_dict)
print("Sum of all items in the dictionary:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

3. **Merge Two Dictionaries:**
```python
def merge_dictionaries(dict1, dict2):
    merged_dict = {**dict1, **dict2}
    return merged_dict

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

result = merge_dictionaries(dict1, dict2)
print("Merged dictionary:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

4. **Convert Key-Values List to Flat Dictionary:**
```python
def convert_to_flat_dictionary(keys, values):
    if len(keys) != len(values):
        return "Number of keys and values should be the same."

    flat_dict = dict(zip(keys, values))
    return flat_dict

keys_list = ['a', 'b', 'c']
values_list = [1, 2, 3]

result = convert_to_flat_dictionary(keys_list, values_list)
print("Flat dictionary:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

5. **Insertion at the Beginning in OrderedDict:**
```python
from collections import OrderedDict

def insert_at_beginning(ordered_dict, key, value):
    ordered_dict.move_to_end(key, last=False)
    ordered_dict[key] = value

my_ordered_dict = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
new_key = 'x'
new_value = 10

insert_at_beginning(my_ordered_dict, new_key, new_value)
print("OrderedDict after insertion at the beginning:", my_ordered_dict)
```

6. **Check Order of Characters in String Using OrderedDict:**
```python
from collections import OrderedDict

def check_order_of_characters(string, pattern):
    char_dict = OrderedDict.fromkeys(pattern)
    j = 0

    for char in string:
        if j == len(pattern):
            break
        if char == pattern[j]:
            j += 1

    return j == len(pattern)

input_string = input("Enter a string: ")
input_pattern = input("Enter a pattern to check: ")

if check_order_of_characters(input_string, input_pattern):
    print("The order of characters in the string matches the pattern.")
else:
    print("The order of characters in the string does not match the pattern.")
```

7. **Sort Python Dictionaries by Key or Value:**
```python
def sort_dictionary_by_key(dictionary):
    return dict(sorted(dictionary.items()))

def sort_dictionary_by_value(dictionary):
    return dict(sorted(dictionary.items(), key=lambda item: item[1]))

my_dict = {'b': 2, 'a': 1, 'd': 4, 'c': 3}

sorted_by_key = sort_dictionary_by_key(my_dict)
sorted_by_value = sort_dictionary_by_value(my_dict)

print("Sorted by key:", sorted_by_key)
print("Sorted by value:", sorted_by_value)
```
