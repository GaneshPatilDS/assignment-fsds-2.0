# ANS:



1. Create a list called `years_list` from your birth year to your fifth birthday:

```python
birth_year = 1980  # Replace with your birth year
years_list = list(range(birth_year, birth_year + 6))
print(years_list)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

2. To find the year when you had your third birthday (you were 0 years old in your first year):

```python
third_birthday_year = years_list[3]
print(third_birthday_year)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

3. To find the year when you were the oldest:

```python
oldest_year = years_list[-1]
print(oldest_year)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

4. Make a list called `things`:

```python
things = ["mozzarella", "cinderella", "salmonella"]
print(things)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

5. Capitalize the element in `things` that refers to a person and then print the list:

```python
things[1] = things[1].capitalize()
print(things)
```

Yes, it changes the element in the list.

6. Make a surprise list:

```python
surprise = ["Groucho", "Chico", "Harpo"]
print(surprise)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

7. Lowercase the last element of the surprise list, reverse it, and then capitalize it:

```python
surprise[-1] = surprise[-1].lower()[::-1].capitalize()
print(surprise)
```

8. Make an English-to-French dictionary called `e2f` and print it:

```python
e2f = {"dog": "chien", "cat": "chat", "walrus": "morse"}
print(e2f)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

9. Write the French word for walrus in your three-word dictionary `e2f`:

```python
french_word_for_walrus = e2f["walrus"]
print(french_word_for_walrus)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

10. Make a French-to-English dictionary called `f2e` from `e2f` using the `items()` method:

```python
f2e = {french: english for english, french in e2f.items()}
print(f2e)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

11. Print the English version of the French word "chien" using `f2e`:

```python
english_word_for_chien = f2e["chien"]
print(english_word_for_chien)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

12. Make and print a set of English words from the keys in `e2f`:

```python
english_words_set = set(e2f.keys())
print(english_words_set)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

13. Make a multilevel dictionary called `life`:

```python
life = {
    "animals": {
        "cats": ["Henri", "Grumpy", "Lucy"],
        "octopi": {},
        "emus": {}
    },
    "plants": {},
    "other": {}
}
print(life)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

14. Print the top-level keys of `life`:

```python
top_level_keys = life.keys()
print(top_level_keys)
```
------------------------------------------------------------------------------------------------------------------------------------------------------

15. Print the keys for `life['animals']`:

```python
animal_keys = life["animals"].keys()
print(animal_keys)
```
------------------------------------------------------------------------------------------------------------------------------------------------------
16. Print the values for `life['animals']['cats']`:

```python
cat_values = life["animals"]["cats"]
print(cat_values)
```
