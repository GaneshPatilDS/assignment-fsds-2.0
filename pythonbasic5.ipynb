ANS:  

       The value of the dictionary key 'foo' with the value 42 would be 42. You can access the value of a          key in a dictionary using the key inside square brackets [] or the get() method. For example:


      my_dict = {'foo': 42}
      print(my_dict['foo'])  # Output: 42

Q.5) If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam      and 'cat' in.keys()?      


ANS: 


     If a dictionary is stored in the variable spam, the expression 'cat' in spam and 'cat' in spam.keys()        are checking for the presence of a key 'cat' in the dictionary, but they are doing it in a slightly          different way.

The expression 'cat' in spam checks if the value 'cat' is present as a key in the dictionary spam. If the key 'cat' is present in the dictionary, the expression returns True, otherwise, it returns False.


     spam = {'cat': 42, 'dog': 56}
     if 'cat' in spam:
     print("The key 'cat' is present in the dictionary.")
     else:
     print("The key 'cat' is not present in the dictionary.")
     
On the other hand, the expression 'cat' in spam.keys() checks if the key 'cat' is present among the list of keys in the dictionary spam. The method .keys() returns a list with the keys of the dictionary.


    if 'cat' in spam.keys():
    print("The key 'cat' is present in the dictionary.")
    else:
    print("The key 'cat' is not present in the dictionary.")
    
Both expressions will return the same output, but using the expression 'cat' in spam is more common and more efficient, as it directly checks if the key is in the dictionary without generating a list of keys first.



Q.6)If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?

ANS:
   
     If a dictionary is stored in the variable spam, the expression 'cat' in spam and 'cat' in spam.values()      are checking for the presence of a value 'cat' in the dictionary, but they are doing it in a slightly        different way.

The expression 'cat' in spam checks if the value 'cat' is present as a key in the dictionary spam. If the key 'cat' is present in the dictionary, the expression returns True, otherwise, it returns False.


    spam = {'cat': 42, 'dog': 56}
    if 'cat' in spam:
        print("The key 'cat' is present in the dictionary.")
    else:
        print("The key 'cat' is not present in the dictionary.")
    
On the other hand, the expression 'cat' in spam.values() checks if the value 'cat' is present among the list of values in the dictionary spam. The method .values() returns a list with the values of the dictionary.


    if 'cat' in spam.values():
        print("The value 'cat' is present in the dictionary.")
    else:
        print("The value 'cat' is not present in the dictionary.")
        
The first expression will return "The key 'cat' is present in the dictionary" , as the key 'cat' is present in the dictionary. While the second expression will return "The value 'cat' is not present in the dictionary.", as there is no value 'cat' in the dictionary, only the key 'cat' which is associated with the value 42.

It's important to note that the first expression is checking the presence of key 'cat' in the dictionary while the second is checking if the value 'cat' exists among all the values in the dictionary.

  
  
  
Q.6)What is a shortcut for the following code?
    if 'color' not in spam: 
    spam['color'] = 'black'
    
ANS:  


   A shortcut for the following code is to use the dict.setdefault(key, default=None) method. This method      returns the value of the key if it is in the dictionary, and if not, it adds the key with the default        value provided and returns the default value.


    spam = {}
    spam.setdefault('color', 'black')
    
This is equivalent to the following code:


    if 'color' not in spam:
        spam['color'] = 'black'
        


    
Q.8) How do you &quot;pretty print&quot; dictionary values using which module and function?

ANS: 

    we can "pretty print" dictionary values using the json module and the json.dumps() function. The            json.dumps() function takes a Python object, such as a dictionary, and converts it to a JSON string,        which is a human-readable format. The json.dumps() method has an optional parameter indent, which when      set to a non-negative integer, will format the JSON output with that many spaces for each level of          indentation.
