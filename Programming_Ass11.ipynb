# ANS:



1. **Find Words Greater Than Given Length:**
```python
def find_words_greater_than_length(sentence, k):
    words = sentence.split()
    result = [word for word in words if len(word) > k]
    return result

sentence = input("Enter a sentence: ")
k = int(input("Enter the value of k: "))
result = find_words_greater_than_length(sentence, k)
print(f"Words greater than length {k} in the sentence: {result}")
```
----------------------------------------------------------------------------------------------------------------------------------

2. **Remove i-th Character from a String:**
```python
def remove_ith_character(input_string, i):
    if i < 0 or i >= len(input_string):
        return "Invalid index."
    return input_string[:i] + input_string[i + 1:]

input_string = input("Enter a string: ")
i = int(input("Enter the index (0-based) of the character to remove: "))
result = remove_ith_character(input_string, i)
print("String after removing the i-th character:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

3. **Split and Join a String:**
```python
def split_and_join(input_string, separator):
    words = input_string.split(separator)
    return " ".join(words)

input_string = input("Enter a string: ")
separator = input("Enter the separator: ")
result = split_and_join(input_string, separator)
print("String after splitting and joining:", result)
```
----------------------------------------------------------------------------------------------------------------------------------

4. **Check if a Given String is Binary String or Not:**
```python
def is_binary_string(input_string):
    return all(char in "01" for char in input_string)

input_string = input("Enter a string: ")
if is_binary_string(input_string):
    print("The given string is a binary string.")
else:
    print("The given string is not a binary string.")
```

5. **Find Uncommon Words from Two Strings:**
```python
def find_uncommon_words(string1, string2):
    set1 = set(string1.split())
    set2 = set(string2.split())
    uncommon_words = set1.symmetric_difference(set2)
    return list(uncommon_words)

string1 = input("Enter the first string: ")
string2 = input("Enter the second string: ")
result = find_uncommon_words(string1, string2)
print("Uncommon words from the two strings:", result)
```

6. **Find All Duplicate Characters in a String:**
```python
def find_duplicate_characters(input_string):
    duplicates = [char for char in input_string if input_string.count(char) > 1]
    return list(set(duplicates))

input_string = input("Enter a string: ")
result = find_duplicate_characters(input_string)
print("Duplicate characters in the string:", result)
```

7. **Check if a String Contains Any Special Character:**
```python
import string

def contains_special_character(input_string):
    special_characters = set(string.punctuation)
    return any(char in special_characters for char in input_string)

input_string = input("Enter a string: ")
if contains_special_character(input_string):
    print("The string contains special characters.")
else:
    print("The string does not contain any special characters.")
```
