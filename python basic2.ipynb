 Q.6) How do you tell the difference between the equal to and assignment operators?Describe a
 condition and when you would use one.
             
             
             
             
    ..The equal to operator (==) is used to compare two values to see if they are equal, while the                 assignment  operator(=) is used to assign a value to a variable. For example, in the following               condition:  
    
       x = 5
       if x == 5:
          print("x is equal to 5")
          
          
          
      Here, the equal to operator is used to compare the value of x to 5. If they are equal, the code block inside the if statement will be executed.

On the other hand, in the following assignment:
    x = 5
    
    
    Here, the assignment operator is used to assign the value of 5 to the variable x.

Q.8)Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints
Greetings! if anything else is stored in spam.    
  
ANS:




     spam = 1
  if spam == 1:
      print("Hello")
  elif spam == 2:
      print("Howdy")
  else:
      print("Greetings!")
 
      In this code, the variable spam is first assigned the value of 1. The code then checks if the value of spam is equal to 1, and if it is, it prints "Hello". If the value of spam is not equal to 1, the code then checks if it's equal to 2 and if it is, it prints "Howdy". If the value of spam is neither 1 nor 2, it will fall in the else block and it will print "Greetings!"
