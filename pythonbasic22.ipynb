# ANS:



1. 
```python
X = 'iNeuron'
def func():
    print(X)

func()
```
Output: 'iNeuron'
Explanation: The variable `X` is defined in the global scope. Inside the function `func()`, it prints the value of `X` which is 'iNeuron'.

------------------------------------------------------------------------------------------------------------------------------------------------------

2. 
```python
X = 'iNeuron'
def func():
    X = 'NI!'

func()
print(X)
```
Output: 'iNeuron'
Explanation: The variable `X` inside the function `func()` is a local variable, separate from the global `X`. It only exists within the function's scope and doesn't affect the global `X`. The print statement outside the function `func()` still refers to the global `X` which remains 'iNeuron'.

------------------------------------------------------------------------------------------------------------------------------------------------------

3. 
```python
X = 'iNeuron'
def func():
    X = 'NI'
    print(X)

func()
print(X)
```
Output: 
```
NI
iNeuron
```
Explanation: Inside the function `func()`, there is a local variable `X` that exists within the function's scope. When `func()` is called, it prints the local `X` which is 'NI'. The print statement outside the function refers to the global `X`, which remains 'iNeuron'.

4. 
```python
X = 'iNeuron'
def func():
    global X
    X = 'NI'

func()
print(X)
```
Output: 'NI'
Explanation: The `global` keyword inside the function `func()` indicates that the variable `X` inside the function refers to the global variable `X`, not a local variable. So, when `func()` is called, it modifies the global `X` to 'NI', and the print statement outside the function reflects the updated value.

5. 
```python
X = 'iNeuron'
def func():
    X = 'NI'
    def nested():
        print(X)
    nested()

func()
print(X)
```
Output: 
```
NI
iNeuron
```
Explanation: Inside the nested function `nested()`, it prints the value of the local `X`, which is 'NI' since it is in the scope of the enclosing function `func()`. The print statement outside the function `func()` refers to the global `X`, which remains 'iNeuron'.

------------------------------------------------------------------------------------------------------------------------------------------------------

6. 
```python
def func():
    X = 'NI'
    def nested():
        nonlocal X
        X = 'Spam'
    nested()
    print(X)

func()
```
Output: 'Spam'
Explanation: The `nonlocal` keyword inside the nested function `nested()` indicates that the variable `X` refers to the variable in the nearest enclosing scope, which is the local variable `X` inside the function `func()`. So, when `func()` is called, the nested function `nested()` modifies the local `X` to 'Spam', and the `print(X)` inside `func()` reflects the updated value.
