

Q.1) What are escape characters, and how do you use them?

ANS:
    
    Escape characters are special characters that are used to represent certain characters or sequences of      characters that are not normally interpreted by the programming language. 


     escape characters is to include a newline character in a string, which is represented by the escap            sequence "\n". For example:

       string = "This is a string\nwith a newline character."

       This would display the string as:

       This is a string
       with a newline character.


Some common escape characters include:

  \n for a newline
  \t for a tab
  \ for a backslash
  \r for a carriage return
  \b for a backspace 
  
It's important to note that the use of escape characters may vary depending on the programming language or system you're using.




Q.2) What do the escape characters n and t stand for?
ANS:
 
    The escape characters "n" and "t" are used in strings to represent a newline and a tab, respectively.
     "n" is used to add a new line after the current line.
     "t" is used to add a tab space before the text.

Q.6) What are the values of the given expressions?
   
   
   "Hello, world!"[1]
   "Hello, world!"[0:5]
   "Hello, world!"[:5]
   "Hello, world!"[3:]


ANS:

     1.The first expression is: 'Hello, world!'[1]. The value of this expression is the character at index 1        of the string, which is "e".

     2.The second expression is: 'Hello, world!'[0:5]. The value of this expression is a substring of the            original string, starting from index 0 and going up to, but not including, index 5. So the value of          this expression is the string "Hello".

     3.The third expression is: 'Hello, world!'[:5]. The value of this expression is also a substring of the        original string, but this time starting from the beginning of the string up to, but not including,            index 5. So the value of this expression is also the string "Hello".

     4.The fourth expression is: 'Hello, world!'[3:]. The value of this expression is a substring of the            original string, starting from index 3 and going to the end of the string. So the value of this              expression is the string "lo, world!".

Please note that these are examples of Python's slicing notation. The behavior may be different in other languages.

Q.8) What are the values of the following expressions?

  1.'Remember, remember, the fifth of July'.split()
  2.'-'.join('There can only one.'.split())

ANS:
   
   1.The first expression is: 'Remember, remember, the fifth of July.'.split(). The split() method splits a string into a list of substrings based on whitespace characters such as spaces, tabs, and newlines. In this case, the value of the expression is a list of substrings: ['Remember,', 'remember,', 'the', 'fifth', 'of', 'July.']

The second expression is: '-'.join('There can only one.'.split()). The first part of the expression is calling split() method on the string "There can only one." which splits it into a list of substrings: ['There', 'can', 'only', 'one.']. The join() method is called on the separator '-' and it joins the elements of the list with the separator. So the value of the expression is a string: 'There-can-only-one.'
